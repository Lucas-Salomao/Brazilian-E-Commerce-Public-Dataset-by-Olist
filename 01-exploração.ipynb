## Exploração básica de algumas tabelas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Função para explorar um dataset
def explore_data(df, name):
    print(f"\nExplorando o dataset {name}:")
    print(df.head())
    print(f"\nInformações sobre {name}:")
    print(df.info())
    print(f"\nEstatísticas descritivas de {name}:")
    print(df.describe())
    print(f"\nValores ausentes em {name}:")
    print(df.isnull().sum())


# Explorar cada dataset
explore_data(customers, "Clientes")
explore_data(geolocation, "Geolocalização")
explore_data(order_items, "Itens do Pedido")
explore_data(order_payments, "Pagamentos")
explore_data(order_reviews, "Avaliações")
explore_data(orders, "Pedidos")
explore_data(products, "Produtos")
explore_data(sellers, "Vendedores")
explore_data(product_category_translation, "Tradução de Categorias")


# Visualizações iniciais (exemplos)
plt.figure(figsize=(10, 5))
sns.countplot(x='order_status', data=orders)
plt.title('Distribuição dos Status dos Pedidos')
plt.show()


plt.figure(figsize=(12, 6))
sns.histplot(order_items['price'], bins=50, kde=True)
plt.title('Distribuição do Preço dos Itens')
plt.show()


# Visualizações adicionais
plt.figure(figsize=(12, 6))
sns.boxplot(x='payment_type', y='payment_value', data=order_payments)
plt.title('Valor do Pagamento por Tipo de Pagamento')
plt.xticks(rotation=45, ha='right') # Rotaciona os labels do eixo x para melhor visualização
plt.show()

plt.figure(figsize=(10, 5))
sns.countplot(x='payment_installments', data=order_payments)
plt.title('Número de Parcelas')
plt.show()


print("\nFim da Exploração Inicial.")
```

```python
# 1. Calcular o total_value em order_items
order_items['total_value'] = order_items['price'] + order_items['freight_value']

# 2. Agregar total_value por order_id
order_totals = order_items.groupby('order_id')['total_value'].sum().reset_index()

# 3. Juntar com o DataFrame orders
orders = pd.merge(orders, order_totals, on='order_id', how='left')


# Agora sim, criar o gráfico de vendas mensais
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
vendas_por_mes = orders.groupby(orders['order_purchase_timestamp'].dt.to_period('M'))['total_value'].sum()

plt.figure(figsize=(12, 6))
vendas_por_mes.plot(kind='line')
plt.title('Evolução das Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Valor Total de Vendas')
plt.show()
```

```python
# 1. Juntar order_items com products para obter a categoria do produto
order_items_with_category = pd.merge(order_items, products[['product_id', 'product_category_name']], on='product_id', how='left')

# 2. Calcular o total vendido por categoria (em valor)
vendas_por_categoria = order_items_with_category.groupby('product_category_name')['price'].sum().sort_values(ascending=False)

# Ou calcular a quantidade vendida por categoria:
# quantidade_por_categoria = order_items_with_category.groupby('product_category_name')['product_id'].count().sort_values(ascending=False)


# 3. Criar o gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=vendas_por_categoria.index, y=vendas_por_categoria.values)  # ou use quantidade_por_categoria
plt.title('Categorias de Produtos Mais Vendidas (em Valor)') # ou 'Categorias de Produtos Mais Vendidas (em Quantidade)'
plt.xlabel('Categoria do Produto')
plt.ylabel('Valor Total Vendido') # ou 'Quantidade Vendida'
plt.xticks(rotation=90, ha='right')  # Rotaciona os labels do eixo x para melhor visualização
plt.tight_layout() # ajusta o layout para evitar que os labels sejam cortados
plt.show()


# Opcional: Mostrar apenas o top N categorias
top_n = 10  # Mostrar as 10 categorias mais vendidas
plt.figure(figsize=(12, 6))
sns.barplot(x=vendas_por_categoria.head(top_n).index, y=vendas_por_categoria.head(top_n).values)
plt.title('Categorias de Produtos Mais Vendidas (em Valor) Top 10') # ou 'Categorias de Produtos Mais Vendidas (em Quantidade)'
plt.xlabel('Categoria do Produto')
plt.ylabel('Valor Total Vendido') # ou 'Quantidade Vendida'
plt.xticks(rotation=90, ha='right')
# # ... (resto do código do gráfico)
```

```python
# 1. Juntar order_items com products
order_items_with_category = pd.merge(order_items, products[['product_id', 'product_category_name']], on='product_id', how='left')

# 2. Calcular o total vendido por categoria
vendas_por_categoria = order_items_with_category.groupby('product_category_name')['price'].sum()

# 3. Calcular o percentual de cada categoria em relação ao total
percentual_vendas = (vendas_por_categoria / vendas_por_categoria.sum()) * 100
percentual_vendas = percentual_vendas.sort_values(ascending=False) # Ordenar por percentual

# Criar o gráfico de barras com os percentuais
plt.figure(figsize=(12, 6))
sns.barplot(x=percentual_vendas.index, y=percentual_vendas.values)
plt.title('Percentual de Vendas por Categoria de Produto')
plt.xlabel('Categoria do Produto')
plt.ylabel('Percentual do Valor Total de Vendas')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()



# Opcional: Adicionar os percentuais no topo de cada barra
# for i, p in enumerate(percentual_vendas):
#     plt.text(i, p, f'{p:.2f}%', ha='center', va='bottom')

# Opcional: Top N categorias
top_n = 10  # Exibir apenas as 10 principais categorias
plt.figure(figsize=(12, 6))
sns.barplot(x=percentual_vendas.head(top_n).index, y=percentual_vendas.head(top_n).values)
plt.title('Categorias de Produtos Mais Vendidas (em Valor) Top 10') # ou 'Categorias de Produtos Mais Vendidas (em Quantidade)'
plt.xlabel('Categoria do Produto')
plt.ylabel('Valor Total Vendido') # ou 'Quantidade Vendida'
plt.xticks(rotation=90, ha='right')
```

```python
# 1. Juntar order_items com products
order_items_with_category = pd.merge(order_items, products[['product_id', 'product_category_name']], on='product_id', how='left')

# 2. Calcular o ticket médio por categoria
ticket_medio_por_categoria = order_items_with_category.groupby('product_category_name')['price'].sum() / order_items_with_category.groupby('product_category_name')['order_id'].nunique()


# 3. Criar o gráfico de barras
plt.figure(figsize=(14, 6))
ticket_medio_por_categoria.plot(kind='bar')
plt.title('Ticket Médio por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Ticket Médio')
plt.xticks(rotation=45, ha='right') # Rotacionar os rótulos do eixo x, se necessário
plt.tight_layout()
plt.show()


print("\nTicket Médio por Categoria:")
print(ticket_medio_por_categoria)
```

```python
# Calcular o ticket médio global
ticket_medio_global = order_items['price'].sum() / len(orders['order_id'].unique())
print(f"Ticket Médio Global: R$ {ticket_medio_global:.2f}")
```

```python
# Agrupa os pedidos por cliente e calcula a soma dos valores totais dos pedidos:
valor_total_por_cliente = orders.groupby('customer_id')['total_value'].sum()

# Agrupa os pedidos por cliente e conta o número de pedidos únicos:
numero_pedidos_por_cliente = orders.groupby('customer_id')['order_id'].nunique()

# Calcula o ticket médio por cliente:
ticket_medio_por_cliente = valor_total_por_cliente / numero_pedidos_por_cliente

# Imprime o resultado:
print("Ticket Médio por Cliente:\n", ticket_medio_por_cliente)

# Para visualizar a distribuição do ticket médio por cliente, você pode usar um histograma:
plt.figure(figsize=(10, 5))
plt.hist(ticket_medio_por_cliente, bins=30) # Define o numero de barras do histograma
plt.title('Distribuição do Ticket Médio por Cliente')
plt.xlabel('Ticket Médio')
plt.ylabel('Número de Clientes')
plt.show()
```

```python
# Converter a coluna de data para datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Juntar order_items com orders para ter o timestamp em order_items
order_items = pd.merge(order_items, orders[['order_id', 'order_purchase_timestamp']], on='order_id', how='left')

# Agrupar por mês e calcular o ticket médio mensal
ticket_medio_mensal = order_items.groupby(order_items['order_purchase_timestamp'].dt.to_period("M"))['price'].sum() / orders.groupby(orders['order_purchase_timestamp'].dt.to_period("M"))['order_id'].nunique()


# Criar um gráfico de linhas com o ticket médio mensal
plt.figure(figsize=(12, 6))
ticket_medio_mensal.plot(kind='line')
plt.title('Ticket Médio Mensal')
plt.xlabel('Mês')
plt.ylabel('Valor do Ticket Médio')
plt.show()

print("\nTicket Médio Mensal:")
print(ticket_medio_mensal)
```
